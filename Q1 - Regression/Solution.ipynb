## Check for missing values
### For Feature cleaning

```python
print 'Column - Uniqe - Null'
for col in X:
    if X[col].isnull().any():
        print col,len(X[col].unique()),X[col].isnull().any(),"--",X[col].isnull().sum()
    else:
        print col,len(X[col].unique()),X[col].isnull().any()
```

## Remove Dirty features

```python
X = X.drop('sold',axis=1)
X = X.drop('desk_id',axis=1)
X = X.drop('hedge_value',axis=1)
X = X.drop('indicator_code',axis=1)
X = X.drop('libor_rate',axis=1)
X = X.drop('portfolio_id',axis=1)
X = X.drop('status',axis=1)
X = X.dropna(subset=['bought'])
```

```python
print 'Column - Uniqe'
for col in X:
    if not(X[col].isnull().any()):
        print col,len(X[col].unique())
    else:
        print col,"--Null--",X[col].isnull().sum()
```

## Convert Labels to Encoded Values

```python
keys = {}
le = LabelEncoder()
#le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
for col in X:
    print col,len(X[col].unique()),X[col].isnull().any()
    if False == X[col].isnull().any() and len(X[col].unique()) < 10:
        X[col] = le.fit_transform(X[col])
        le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
        keys[col] = le_name_mapping
        #print "Yup"
```

## Perform same cleaning on the test set as the train set

```python
#data_test = pd.read_csv('test.csv')
data_test = data_test.drop('portfolio_id',axis=1)
data_test = data_test.drop('desk_id',axis=1)
data_test = data_test.drop('sold',axis=1)
data_test = data_test.drop('libor_rate',axis=1)
data_test = data_test.drop('indicator_code',axis=1)
data_test = data_test.drop('hedge_value',axis=1)
data_test = data_test.drop('status',axis=1)

#data_test = data_test.drop('return',axis=1)
```

```python
data_test.head()
```

## Function to map keys to labels

```python
def convert2(x,col):
    return keys[col][x]

for col in data_test:
    #print(X[col])
    if col in keys.keys():
        data_test[col] = [convert2(a,col) for a in data_test[col]]
```

## Plot the train set on the trained Regressor
